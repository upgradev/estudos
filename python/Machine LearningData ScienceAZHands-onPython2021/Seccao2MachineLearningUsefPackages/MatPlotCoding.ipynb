```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
Temp_I = [0.72, 0.61, 0.65, 0.68, 0.75, 0.90, 1.02, 0.93, 0.85, 0.99, 1.02]

plt.plot(Year, Temp_I)
plt.xlabel("Year")
plt.ylabel("Temp_I")
plt.title("Global Warming", {"fontsize": 12, 'horizontalalignment': 'center'})
plt.show()
```

```python
Month = ["Jan", "Feb", "March", "April", "May",
         "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]
Customer_1 = [12, 13, 9, 8, 7, 8, 8, 7, 6, 5, 8, 10]
Customer_2 = [14, 16, 11, 7, 6, 6, 7, 6, 5, 8, 9, 12]

plt.plot(Month, Customer_1, color = "red", label="Costumer 1", marker = "o")
plt.plot(Month, Customer_2, color = "blue", label="Costumer 2", marker = "^")
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Building Consumption")
plt.legend(loc="upper right") #label plot
plt.show()
```

```python
plt.subplot(1,2,2)
plt.plot(Month, Customer_1, color = "red", label="Costumer 1", marker = "o")
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Building Consumption")
plt.legend(loc="upper right") #label plot
plt.show()

plt.subplot(1,2,2)
plt.plot(Month, Customer_2, color = "blue", label="Costumer 2", marker = "^")
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Building Consumption")
plt.legend(loc="upper right") #label plot
plt.show()
```

```python
plt.scatter(Month, Customer_1, color="red", label="Customer 1")
plt.scatter(Month, Customer_2, color="blue", label="Customer 2")
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Scatter Plot of building consumption")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.hist(Customer_1, bins=20, color="green")
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Histogram")
plt.show()
```

```python
plt.bar(Month, Customer_1, width=0.8, color="blue")
plt.show()
```

```python
plt.bar(Month, Customer_1, color="red", label="Customer 1")
plt.bar(Month, Customer_2, color="blue", label="Customer 2")
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Bar Plot of building consumption")
plt.legend()
plt.show()
```

```python
bar_width = 0.4
Month_b = np.arange(12)

plt.bar(Month_b, Customer_1, bar_width, color="blue", label="Customer 1")
plt.bar(Month_b+bar_width,Customer_2, bar_width, color="red", label="Customer 2")
plt.title("Bar Charts")
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.xticks(Month_b + (bar_width) / 12, ("Jan", "Feb", "March", "April", "May",
                                        "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"))
plt.legend()
plt.show()
```

```python

plt.boxplot(Customer_1, notch=True, vert=True)
plt.boxplot([Customer_1, Customer_2], 
            patch_artist=True,
            boxprops=dict(facecolor="red", color="blue"), 
            whiskerprops=dict(color="green"), 
            capprops=dict(color="yellow"), 
            medianprops=dict(color="brown"))
plt.show()
```
