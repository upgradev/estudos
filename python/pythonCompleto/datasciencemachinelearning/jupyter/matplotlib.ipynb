# Primeiro plot

```python
x = [5,10,15,20,25]
y = [15, 25, 45, 55, 60]
plt.plot(x, y)
```

```python
x = [5,10,15,20,25]
y = [15, 25, 45, 55, 60]
fig,axs = plt.subplots(figsize=(12,4))
axs.set_xlabel("Eixo x")
axs.set_ylabel("Eixo Y")
axs.set_title("titulo do gráfico")
axs.plot(x, y, "ro--", label="Dados 1")
axs.plot(y, x, "g<--", label="Dados 2")
axs.legend() #mostrar a legenda
```
